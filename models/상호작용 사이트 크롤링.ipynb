## 일산병원 데이터 수집

```python
driver = webdriver.Chrome('./chromedriver.exe')
driver.implicitly_wait(3)
```

```python
data_list = []

for i in tqdm(range(2, 53)):
    url = 'https://www.nhimc.or.kr/board/disFdBoardView.do?page=1&no='+ str(i)
    
    response = requests.get(url)  
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # 제목 가져오기
    title = driver.find_element_by_xpath('//*[@id="contents"]/article/section/div/div[3]/h3')
    title = [title]

    # 소제목 가져오기
    element1 = soup.find_all('h4', class_='cnt_tit tit_icon')
    title2 = []
    for ele in element1:
        text = ele.find('span').get_text()
        title2.append(text)
    title2

    # 본문 가져오기
    element2 = soup.find_all('div', class_='txt')
    text_list = []
    for ele in element2:
        text = ele.text
        text_list.append(text)
    text_list

    for title2, text in zip(title2, text_list):
        data_list.append([title[0], title2, text])
```

```python
df = pd.DataFrame(data_list)

# 특정 컬럼에서 \r, \n, \t 제거하는 함수
def clean_text(text):
    return text.replace('\r', '').replace('\n', '').replace('\t', '')

# '2' 컬럼의 공백 문자 제거
df['2'] = df['2'].apply(clean_text)
```

```python
df.to_csv('일산병원_질병.csv')
```

## 서울아산병원

```python
data_list = []

for i in tqdm(range(18, 122)):
    url = 'https://www.amc.seoul.kr/asan/healthinfo/mealtherapy/mealTherapyDetail.do?pageIndex=11&searchCondition=&searchKeyword=&mtId=' + str(i)
    response = requests.get(url)  
    soup = BeautifulSoup(response.text, 'html.parser')
    
    #질병명
    disease = soup.find('strong', class_='contTitle').text
    disease = [disease]

    # 소제목 가져오기
    title2 = []
    element1 = soup.find_all('dt', class_='dep2Title')
    for title in element1:
        title = title.text
        title2.append(title)
    title2 = title2[:-1]

    # 내용 가져오기
    text_list = []
    desc_dl = soup.find('dl', class_='descDl')
    p_tags = desc_dl.find_all('p')
    for tag in p_tags:
        text = tag.get_text(strip=True)
        text_list.append(text)
        text_list = [item for item in text_list if item != '']

    for title2, text in zip(title2, text_list):
        data_list.append([disease[0], title2, text])
```

```python
data_list
```

```python
df = pd.DataFrame(data_list)
df
```

```python
df.to_csv('서울아산병원_질병.csv')
```

## 세브란스 병원

```python
driver = webdriver.Chrome('./chromedriver.exe')
driver.implicitly_wait(3)
```

```python
# 웹페이지를 불러옵니다.
driver.get('https://health.severance.healthcare/health/lifecare/nutrition/diseasediet.do')

# Find the element using XPath
try:
    while True:
        button = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, '//*[@id="btnMoreArticle"]'))
        )

        button.click()
except:
    print()

# div 태그 안에 있는 href 속성을 가진 요소를 찾습니다.
url_list = []
div_elements = driver.find_elements_by_css_selector('div.thumb-item > a[href]')
for element in div_elements:
    href_value = element.get_attribute('href')
    url_list.append(href_value)
```

```python
url = 'https://health.severance.healthcare/health/lifecare/nutrition/diseasediet.do?&mode=view&articleNo=109035&articleLimit=0'
response = requests.get(url, verify=False)  
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
data_list = []

for url in tqdm(url_list):
    response = requests.get(url, verify=False)  
    soup = BeautifulSoup(response.text, 'html.parser')

    # 질병명 찾기
    disease = soup.find('h3', class_='subject').text
    disease = [disease]

    # 소제목 찾기
    title2 = soup.find('div', class_='fr-view')
    title2 = title2.find_all('strong')

    title_list = []
    for title in title2:
        title_list.append(title.text)

    # 본문 가져오기
    fr_view = soup.find('div', class_='fr-view')
    fr_view = fr_view.find_all('p')

    text_list = []
    for text in fr_view:
        text = text.text
        text_list.append(text)

    split_indices = [i for i in range(0, len(text_list)-2) if text_list[i + 1] == '' and text_list[i + 2] == '']
    split_indices.append(len(text_list)-1)

    num = 0
    text_list_cleaned = []
    text_list2 = []

    for i in text_list: 
        text_list2.append(i)
        if num in split_indices:
            text_list_cleaned.append(text_list2)
            text_list2 = []
        num+=1
    
    for title2, text in zip(title2, text_list_cleaned):
        data_list.append([disease[0], title2, text])
```
