# filtered_rows13의 같은 성분을 한 행에 , 으로 join한게 combined_text

```python
combined_text = filtered_rows13.groupby(['sunb_name', 'ingd_code'])['contents_origin'].apply(lambda x: ' '.join(x)).reset_index()
```

```python
combined_text
```

# 권고사항이 있는 영양소들 ---> 11797

권고사항이 있는 영양소들 추출한게 matching_info_rows

```python
matching_info_rows = ingd[ingd['ingd_code'].isin(combined_text['ingd_code'])]
```

```python
matching_info_rows
```

# 부작용이 있는 약들의 모임 matching_med_rows

권고사항이 있는 영양소들을 가진 약들의 목록 matching_med_rows

```python
matching_med_rows = filtered_med[filtered_med['dcode'].isin(matching_info_rows['dcode'])]
```

```python
matching_med_rows
```

# matching_med_rows 약에  해당하는 ingd_code 컬럼 추가하기

하나의 약이지만 여러가지 성분을 가진 약들이 있기 때문에 행이 늘어났다.

```python
merged_df = pd.merge(matching_med_rows, ingd[['dcode', 'ingd_code']], on='dcode', how='left')
```

```python
merged_df
```

# 늘어난 행들이 있으니까 dcode가 같으면 ingd_code를 한 행에 넣기

7898--> 21446 ---> 7898

```python
merged_df['ingd_code'] = merged_df.groupby('dcode')['ingd_code'].transform(lambda x: ', '.join(x.dropna()))

merged_df2 = merged_df.drop_duplicates(subset='dcode')
```

```python
merged_df2
```

```python
combined_text
```

```python
merged_df2['merged_contents'] = merged_df2['ingd_code'].apply(
    lambda codes: ', '.join(combined_text.loc[combined_text['ingd_code'].astype(str).isin(map(str, codes.split(','))), 'contents_origin'])
)
```
